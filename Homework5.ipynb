```python
#Problem 1:
# Given a positive integer n (assume n>=2), make a list of its divisors, starting with 2 and 
#ending with n itself

def factors(n):
    if n == 0 or n == 1:
        return None 
    divisor = []
    for i in range(2,n+1):
        if n % i == 0:
            divisor.append(i)
    return divisor
print factors(2), factors(3), factors(6)
print factors(7), factors(9), factors(12)
```

```python
#Problem 2
#using factors(), return True if n is prime

def isPrime(n):
    divisor = factors(n)
    if len(divisor) == 1:
        return True
    if len(divisor) != 1:
        return False 
print isPrime(2), isPrime(3), isPrime(6)
print isPrime(7), isPrime(9), isPrime(12)
```

```python
#problem 3
# Now let's count the number of times each factor of n divides d.
# For example, 45=3*3*5 so  howManyTimesDivides(45, 3)=2 because we encounter 3 twice.

# hint: this is easy with recursion, like we did the Factorial and Fibonacci functions

def howManyTimesDivides(n, d):
    if not n % d == 0:
        return 0
    else:
        return 1 + howManyTimesDivides (n/d, d)    
    
print(howManyTimesDivides(2,2))
print(howManyTimesDivides(7,2))

print(howManyTimesDivides(45,3))
print(howManyTimesDivides(45,5))

for i in range(2,13):
    print i, howManyTimesDivides(360,i)
```

```python
#problem 4:
# Make a dict object with the counts of each factor of n like the examples below
# Use the functions you've already defined to help you

def factorCounts(n):
    result = {}
    divisors = factors(n)
    for divisor in divisors:
        result[divisor] = howManyTimesDivides(n, divisor)
    return result

#Compare your outputs - note that entries in dict objects are not always in the same order

print(factorCounts(2))
print(factorCounts(3))
print(factorCounts(12))
print(factorCounts(360))
```

```python
#problem 5:
# now use the factorCounts object you've just constructed
# to write out (using the print command in your function)
# the table of _prime_ factor counts following the pattern below
#hint: you'll want to use the rjust function to align your extra spaces
#  example: str(1).rjust(5), str(123).rjust(5)

def printTable(n):
    result = factorCounts(n)
    divisors = factors(n)
    for i in range(2, n+1): 
        if isPrime(i) == True:
            if i in divisors:
                print str(i).rjust(5), '*' * result[i]
            else:
                print str(i).rjust(5), '-'
    
```
